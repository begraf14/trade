```python
def get_pending_order(symbol):
    pending = mt5.orders_get(symbol=symbol)
    
    if pending:
        pending = pd.DataFrame(list(pending),columns=pending[0]._asdict().keys())
        return pending    
    else:
        return pending
```

```python
def cancel_pending_order(order_number , symbol ):
    # Create the request
    request = {
        "action": mt5.TRADE_ACTION_REMOVE,
        "order": order_number,
        "comment": "Order Removed"
    }
    # Send order to MT5
    order_result = mt5.order_send(request)
    

    if order_result[0] != 10009:
        print(f"Error placing order. ErrorCode {order_result[0]}, Error Details: {order_result}")

    return order_result
```

```python
def get_positions(symbol):
    pos = mt5.positions_get(symbol=symbol)
    
    if pos : 
        pos =pd.DataFrame(list(pos),columns=pos[0]._asdict().keys())  
        return pos
    
    else:
        return pos    
```

```python
def place_order(order_type, symbol, volume, price, stop_loss, take_profit, comment):
    # If order type SELL_LIMIT
    if order_type == 3:
        order_type = mt5.ORDER_TYPE_SELL_LIMIT
    elif order_type == 2:
        order_type = mt5.ORDER_TYPE_BUY_LIMIT
    # Create the request
    request = {
        "action": mt5.TRADE_ACTION_PENDING,
        "symbol": symbol,
        "volume": volume,
        "type": order_type,
        "price": round(price, 3),
        "sl": round(stop_loss, 3),
        "tp": round(take_profit, 3),
        "type_filling": mt5.ORDER_FILLING_RETURN,
        "type_time": mt5.ORDER_TIME_GTC,
        "comment": comment
    }
    # Send the order to MT5
    order_result = mt5.order_send(request)
    # Notify based on return outcomes
    if order_result[0] != 10009:
        print(f"Error placing order. ErrorCode {order_result[0]}, Error Details: {order_result}")

    return order_result
```

```python
def pivot_points(symbol):
    last = (datetime.datetime.now() - datetime.timedelta(hours=1))

    m1 = mt5.copy_rates_from(symbol, mt5.TIMEFRAME_D1, last, 2)[0]
    
    high = m1[2]
    low =  m1[3]
    close = m1[4]
    range_ = high - low

    pp = (high+ low + close)/3

    r1 = 2*pp - low
    r2 = pp + range_
    r3 = pp + range_*2
    r4 = pp + range_*3
    r5 = pp + range_*4 

    s1 = 2*pp - high
    s2 = pp - range_
    s3 = pp- range_*2
    s4 = pp- range_*3
    s5 = pp- range_*4
    
    liste= np.array([pp , r1, r2, r3, r4 ,r5, s1, s2, s3, s4, s5])
    
    return liste    
```

```python
# Function to modify an open position
def modify_position(order_number, symbol, new_sl, new_tp):
    # Create the request
    request = {
        "action": mt5.TRADE_ACTION_SLTP,
        "symbol": symbol,
        "sl": new_sl,
        "tp": new_tp,
        "position": order_number
    }
    # Send order to MT5
    order_result = mt5.order_send(request)
    if order_result[0] == 10009:
        return print(f"{symbol} position updated successful")
    else:
        return order_result
```
